```
%tensorflow_version 2.x
from tensorflow.keras import *
from tensorflow.keras.layers import *
import numpy as np

from google.colab import files
uploaded = files.upload()
#BrainBody.txt 파일 불러오기
data = np.genfromtxt('BrainBody.txt', encoding='ascii')
#column 0 (index) 삭제
data = np.delete(data, np.s_[0:1], axis=1)
#column split
dataX, dataY = np.split(data, 2, axis = 1)
```

```
num_test = int(dataX.shape[0] * 0.2)
num_train = dataX.shape[0] - num_test

trainX = dataX[:num_train]
trainY = dataY[:num_train]
testX = dataX[num_train:]
testY = dataY[num_train:]
```

```
model = Sequential()
model.add(Dense(10, input_shape=(dataX.shape[1],)))
model.add(Activation('relu'))
model.add(Dense(10))
model.add(Activation('relu'))
model.add(Dense(1))
model.summary()
```

```
sgd = optimizers.SGD(lr = 0.01)
model.compile(optimizer=sgd, loss = 'mean_squared_error', metrics = ['mse'])
```

```
model.fit(trainX, trainY, batch_size=16, epochs = 500, verbose=1)
```

```
loss, mse = model.evaluate(testX, testY)
```
